### 1.1. Скачать следующие данные: kc-house-data и laptop_price


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("kc_house_data.csv", sep = ",", encoding = "utf-8")
pd.set_option('display.max_columns', None)
print(df.head())
```

### 1.2. Изучите стоимости недвижимости
**Вывод: подавляющее большинство домов стоит в пределах 1-1,5 млн**

```python
plt.figure(figsize = (8, 6))
plt.hist(df["price"])
plt.title("Cтоимость недвижимости")
plt.xlabel("Цена, млн")
plt.ylabel("Количество домов")
```

### 1.3. Изучите распределение квадратуры жилой
**Вывод: подавляющее большинство домов имеют площадь жилых помещений до 3000 м2, небольшая часть - от 3000 до 4000 м2, больше - единичные случаи.**

```python
plt.figure(figsize = (8, 6))
plt.hist(df["sqft_living"])
plt.title("Квадратура жилая")
plt.xlabel("м2")
plt.ylabel("Количество")
```

### 1.4. Изучите распределение года постройки
**Вывод: более половины домов в выборке построены во второй половине 20 века; на втором месте по количеству - построенные в 21 веке; и ненамного меньше них - домов постройки первой половины 20 века.**

```python
plt.figure(figsize = (8, 6))
plt.hist(df["yr_built"])
plt.title("Год постройки дома")
plt.xlabel("год")
plt.ylabel("Количество")
```

### 2.1. Изучите распределение домов от наличия вида на набережную

### 2.2. Изучите распределение этажей домов
**Вывод: только 3,7% домов имеют более 2 этажей. Самая распространенная этажность - 1 этаж (49,4%); вторая по распространенности - 2 этажа (38,1%). Обусловлено может быть факторами: цена (одноэтажные менее дороги), достаточность жилого пространства для комфорного проживания (2 этажа - прилично площади) А вот этажность 1,5 сравнительно редка (8,8%): видимо, этаж 0,5 тесен и потому малопригоден под полезные нужды.**

```python
plt.figure(figsize = (12, 8))
floor = df["floors"].value_counts()
plt.pie(floor, autopct='%1.1f%%')
plt.title("Этажность")
plt.legend(floor.index)
```

### 2.3. Изучите распределение состояния домов
**Вывод: почти 2/3 (64,3%) домов - в состоянии "на троечку" в буквальном смысле (состояние 3). Чуть более 1/3 домов - в хорошем и совсем хорошем состоянии (состояние 4 26,7%, состояние 5 7,7%). Рухляди почти нет - менее 1% выборки.**

```python
plt.figure(figsize = (12, 8))
houses_condition = df["condition"].value_counts()
plt.pie(houses_condition, autopct='%1.1f%%')
plt.title("Состояние дома")
plt.legend(houses_condition.index)
```

Экспериментальная выборка: столбец price (тип данных float64)
Какие столбцы могут выступать в роли влияющих на него факторов: bedrooms  bathrooms  sqft_living sqft_lot  floors  waterfront  view  condition  grade  sqft_above sqft_basement  yr_built  yr_renovated

```python
print(df.info())
```

Что точно входит в стоимость нового дома?
1. Затраты на его постройку: они находятся в зависимости от столбцов bedrooms  bathrooms  sqft_living sqft_lot  floors sqft_above sqft_basement
2. Удобство расположения: этот фактор нельзя оценить по данной таблице без подтягивания картографических данных - однако даже и с ними, не зная множества иных, уже связанных с географическим расположением дома, факторов, дать оценку невозможно.
3. Окружение дома: waterfront  view

Что учитывается для старого дома?
1. Насколько он старый: yr_built
2. Общее текущее состояние: condition
3. Ремонтировался ли и как давно: yr_renovated

столбец grade содержит, предположительно, некоторый интегральный качественный показатель: из чего и как именно он складывается, неясно (но, скорее всего, из всех остальных параметров)

На что смотрит покупатель при осмотре дома?
1. Общее текущее состояние: condition
2. Сколько в нем комнат. В данной таблице о количестве именно комнат (а не только спален или ванных) можно судить по: жилой площади sqft_living и количеству этажей floors, ну и по спальням и ванным, конечно, тоже - bedrooms  bathrooms
3. Размер придомового участка и/или парковки: sqft_lot
4. Окружение дома: view (а waterfront скорее является специфической частью view, чем самостоятельной характеристикой, т.к. всего 0,8% домов выборки им обладают)

*Гипотеза 1.* Цена дома прямо связана с общим текущим состоянием. 
*Результат:* в сторону возрастания цены - да, только для домов с низким общим состоянием. При состоянии 3 и выше связи не видно: точки группируются в одном и том же ценовом диапазоне для всех состояний.
*Вывод*: Общее текущее состояние оказывает значимое влияние на цену только для домов-развалюшек.

```python
sns.jointplot(x = df["condition"], y = df["price"])
```

Прежде чем проверять гипотезу 2 о цене, надо посмотреть на взаимосвязи между признаками, описывающими габариты дома, чтобы выбрать самый интегральный из них и его уже применить для проверки гипотезы 2 (чтобы избежать множественных сравнений, чреватых ошибками)

*Малая гипотеза 1.* Жилая площадь прямо связана с количествами спален и ванных.
*Вывод*: со спальнями - нет; с ванными - положительная связь, быстро возрастает по параболе. Вероятно, площадь ванных составляет заметную долю жилой площали.
Жилая площадь имеет почти нормальное распределение, лишь немного сдвинутое в меньшую сторону.
Спальни - нормальное распределение. Ванные - распределение значимо сдвинуто в меньшую половину: вероятно, действует фактор отбора (вряд ли в доме нужно много ванных + почти в половине домов выборки всего 1 этаж, а там ванных немного)

```python
sns.jointplot(x = df["sqft_living"], y = df["bedrooms"])
```

```python
sns.jointplot(x = df["sqft_living"], y = df["bathrooms"])
```
